```python
df = df.withColumn('age', col('age').cast('int')) \
    .withColumn('experience', col('experience').cast('int')) \
    .withColumn('salary', col('salary').cast('int'))
df.printSchema()
```

```python
df.show(7)
```

```python
df.columns
```

```python
from pyspark.ml.feature import VectorAssembler

featureAssembler = VectorAssembler(
    inputCols= ['age', 'experience'],
    outputCol= 'Independent Variable'
)
out = featureAssembler.transform(df)
out.show()
```

```python
df_model = out.select('Independent Variable', 'salary')
df_model.show()
```

```python
from pyspark.ml.regression import LinearRegression

train_data, test_data = df_model.randomSplit([0.8,0.2])
lr = LinearRegression(featuresCol='Independent Variable', labelCol='salary')
```

```python
lr = lr.fit(train_data)
```

```python
lr.coefficients
```

```python
lr.intercept
```

```python
pred_out = lr.evaluate(test_data)
pred_out.predictions.show()
```

```python
(pred_out.meanAbsoluteError, pred_out.meanSquaredError, pred_out.rootMeanSquaredError)
```
